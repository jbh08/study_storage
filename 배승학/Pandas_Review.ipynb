### Manipulating Dataframe

#### select columns : 특정 column을 선택하는 방법은 크게 3가지가 있다. 
    1. attribute 방식 : data.column_name
    2. key 방식 : data['column_name']
    3. indexing 방식 : data.iloc[:, <column_number>]
    4. filter 함수 사용하기 : data.filter(regex = '')
    5. data type을 사용하여 구하기 : select_dtypes('object')
    
#### slicing하기. 
- 하나의 행을 알아보는 방법 : iloc, loc
    * iloc() :  iloc는 마지막이 포함되지 않는다. 코마 indexing을 지원한다. python기반과 똑같은 기법이다. 
    * loc() : loc는 마지막이 포함된다. 
    
    
- 특정값 데이터를 뽑아오는 방법 : iloc, loc, iat, at
    * iat() : iat는 숫자방식.
    * at() : 이름방식
    
#### Row나 column 삭제하기. (drop)
- row나 column을 삭제하기 위해서 drop()함수를 사용한다. 
- drop 함수에서 axis를 1로 설정하면 column을 0으로 설정하면 row를 삭제한다. 

#### column 이름 rename
- rename 함수를 사용한다. rename({columns = old_column_name : new_column_name})

# Missing Value 처리하기.

missing value가 존재하면 missing value를 처리해야 한다. mssing value를 처리하기 위해선 크게 2가지 방법이 있다. 

### 1. 없애버리기.


- 만약에 missing value가 별로 없다면 없애버릴 수 있다. 
- dropna() 함수를 사용한다. 


### 2. 특정값 넣기.
혹은 missing value에 아래와 같은 값을 넣을 수 있다. 


 - 전체 평균값을 넣는다 .
 
 - 비슷한 값을 가진 데이터의 평균을 넣는다. 
 
 - KNN을 써서 예측값을 넣는다.
 
* coldstart : data가 부족하여 알고리즘이 잘 학습되지 않는 상태.

### missing value 없애버리기. 

```python
# dropna()함수를 사용하이 없애기. 
mpg.dropna(inplace=True)
```

```python
# na인 value가 없는 것을 볼 수 있다. 
mpg[mpg.horsepower.isna()]
```

```python
mpg = mpg_raw.copy()
```

### 특정값 넣기.

- 전체 평균 넣는 것
- 비슷한 애들 평균 넣는 것.
- KNN 써서 예측해서 넣는 것. 

```python
mpg[mpg.horsepower.isna()]
```

```python
# 0으로 대체해보기
mpg[mpg.fillna(0).horsepower.isna()]
```

```python
# 평균값으로 대체하기.
mpg.fillna(mpg.mean())
```

# Tidy Data

데이터 분석을 수행하면서 다양한 데이터 변환 작업을 해야 한다. 애초에 데이터가 특정 분석을 위해 만들어지는 경우가 없기 때문에, 데이터분석을 위해 데이터 변환 및 전처리 작업을 진행해주어야 한다. 

## tidy data란?
위키피디아에 따르면 Tiday data의 정의는 아래와 같다. 
"Tidy data there's no need to start from scratch"
즉, 데이터 처링 가장 많은 시간이 소요되는 전처리 과정을 줄이기 위해 처음부터 데이터를 표준화해서 만들자는 것이다. 처음부터 기본적인 규칙을 가지고 데이터를 만들면 시간을 아낄 수 있을 것이다. 

## Tiday data의 특징
1. 각 변수는 개별의 열(column)으로 존재한다. 


2. 각 관측치는 행(row)를 구성한다. 


3. 각 표는 단 하나의 관측기준에 의해서 조직된 데이터를 저장한다. 


4. 만약 여러개의 표가 존재한다면, 적어도 하나이상의 열(column)이 공유되어야 한다. 
    - 변수(Variable) : 키, 몸무게, 성별 등
    - 값(Value) : 152cm, 50kg, 여성
    - 관측치(Observation) : 값을 측정한 단위 (여기서는 각각의 사람)
    

### 지저분한 데이터의 예:
Name          Treatment A  Treatment B

John Smith         -	       2

Jane Doe	       16	       11

Mary Johnson	   3	       1

### 깔끔한 데이터(Tidy data)의 예:
Name	    Treatment	Result
John Smith	   a	      -
Jane Doe	   a	      16
Mary Johnson   a	      3
John Smith	   b	      2
Jane Doe	   b	      11
Mary Johnson   b	      1


## 지저분한 데이터의 일반적인 모습
1. 열이름(Column header)이 변수 이름이 아니고 값인 경우


2. 같은 표에 다양한 관측 단위(observational units)가 있는 경우


3. 하나의 열(column)에 여러 값이 들어있는 경우


4. 변수가 행과 열에 모두 포함되어 있는 경우


5. 하나의 관측 단위(observational units)가 여러 파일로 나뉘어져 있는 경우.
